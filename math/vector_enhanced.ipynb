## 1. Vector Basics

### Definition
A vector is a quantity with both **magnitude** (size) and **direction**.

### üåç Real-World Example: GPS Navigation
When your GPS says "Go 5 miles northeast," that's a vector! The magnitude is 5 miles, and the direction is northeast. Without both pieces of information, you'd be lost!

### Representation
- **Component form**: v = (x, y) or v = (x, y, z)
- **Unit vector notation**: v = xi + yj + zk

### üéØ Use Cases:
- **Gaming**: Character movement (speed + direction)
- **Weather**: Wind velocity (speed + direction)
- **Physics**: Force, velocity, acceleration
- **Computer Graphics**: Object positioning and movement

```python
# Creating vectors - like setting coordinates on a map
v1 = np.array([3, 4])  # Move 3 units right, 4 units up
v2 = np.array([1, -2, 3])  # 3D position: x=1, y=-2, z=3

print(f"2D vector (like a GPS coordinate): {v1}")
print(f"3D vector (like a drone position): {v2}")
```

## 2. Vector Magnitude (Length)

**Formula**: ||v|| = ‚àö(x¬≤ + y¬≤ + z¬≤)

### üö∂ Real-World Example: Distance Calculation
Imagine you walk 3 blocks east and 4 blocks north from your starting point. The magnitude tells you the straight-line distance from where you started (5 blocks using the Pythagorean theorem). This is like "as the crow flies" distance!

### üéØ Use Cases:
- **Engineering**: Calculating the strength of forces
- **Computer Graphics**: Distance between objects for collision detection
- **Sports**: Ball speed regardless of direction
- **Navigation**: Straight-line distance calculations

```python
def magnitude(v):
    return np.sqrt(np.sum(v**2))

# Example: You walk 3 blocks east, 4 blocks north
walk_vector = np.array([3, 4])
distance_from_start = magnitude(walk_vector)

print(f"You walked in direction: {walk_vector}")
print(f"Straight-line distance from start: {distance_from_start} blocks")
print(f"This is the same as: {np.linalg.norm(walk_vector)} blocks")
```

## 3. Unit Vectors

A unit vector has magnitude 1. **Formula**: √ª = v/||v||

### üß≠ Real-World Example: Compass Directions
A unit vector is like a compass needle - it only shows direction, not distance. "North" is always "1 unit north" regardless of how far you want to go. It's pure direction information!

### üéØ Use Cases:
- **3D Graphics**: Surface normals for realistic lighting calculations
- **Robotics**: Direction commands ("turn this way")
- **Navigation**: Pure directional information
- **Physics**: Representing directions without magnitude

```python
def unit_vector(v):
    return v / np.linalg.norm(v)

# Example: Convert any movement to just direction
movement = np.array([6, 8])  # Walk 6 east, 8 north
direction_only = unit_vector(movement)

print(f"Original movement: {movement}")
print(f"Direction only (unit vector): {direction_only}")
print(f"Magnitude check: {np.linalg.norm(direction_only)} (should be 1.0)")
print(f"This tells us to go {direction_only[0]:.2f} east for every {direction_only[1]:.2f} north")
```

## 4. Vector Addition and Subtraction

- **Addition**: v + w = (v‚ÇÅ + w‚ÇÅ, v‚ÇÇ + w‚ÇÇ, v‚ÇÉ + w‚ÇÉ)
- **Subtraction**: v - w = (v‚ÇÅ - w‚ÇÅ, v‚ÇÇ - w‚ÇÇ, v‚ÇÉ - w‚ÇÉ)

### ‚úàÔ∏è Real-World Example: Airplane Navigation
If a plane flies east at 200 mph and there's a 50 mph wind from the south, vector addition gives the plane's actual path and speed relative to the ground. The pilot needs to know both the intended flight vector AND the wind vector to reach the destination!

### üéØ Use Cases:
- **Physics**: Combining forces (gravity + thrust + drag)
- **Gaming**: Character movement with environmental effects (walking + conveyor belt)
- **Engineering**: Structural load analysis
- **Economics**: Combining different market forces

```python
# Airplane example
plane_velocity = np.array([200, 0])  # 200 mph east
wind_velocity = np.array([0, 50])    # 50 mph from south (pushing north)

actual_velocity = plane_velocity + wind_velocity
actual_speed = np.linalg.norm(actual_velocity)

print(f"Plane intends to fly: {plane_velocity} mph")
print(f"Wind is blowing: {wind_velocity} mph")
print(f"Actual flight path: {actual_velocity} mph")
print(f"Actual ground speed: {actual_speed:.1f} mph")

# Simple walking example
step1 = np.array([2, 3])  # Walk 2 east, 3 north
step2 = np.array([1, -1]) # Then 1 east, 1 south
total_displacement = step1 + step2

print(f"\nFirst step: {step1}")
print(f"Second step: {step2}")
print(f"Total displacement: {total_displacement}")
```

## 5. Scalar Multiplication

**Formula**: kv = (kv‚ÇÅ, kv‚ÇÇ, kv‚ÇÉ)

### üç≥ Real-World Example: Scaling Recipes
If a recipe calls for ingredients in certain proportions and you want to make 3 times as much, you multiply each ingredient by 3. Same direction (same recipe), but bigger magnitude (more food)!

### üéØ Use Cases:
- **Computer Graphics**: Scaling objects (making them bigger/smaller)
- **Physics**: Amplifying forces (2x the force = 2x the acceleration)
- **Economics**: Scaling production or investment
- **Engineering**: Proportional scaling of designs

```python
# Recipe scaling example
original_recipe = np.array([2, 1, 3])  # 2 cups flour, 1 cup sugar, 3 eggs
scale_factor = 3  # Make 3x as much
scaled_recipe = scale_factor * original_recipe

print(f"Original recipe: {original_recipe} (flour, sugar, eggs)")
print(f"Scaled by {scale_factor}: {scaled_recipe}")

# Physics example - force scaling
force = np.array([10, 5])  # 10N east, 5N north
double_force = 2 * force
half_force = 0.5 * force

print(f"\nOriginal force: {force} N")
print(f"Double the force: {double_force} N")
print(f"Half the force: {half_force} N")
```

## 6. Dot Product (Scalar Product)

**Formula**: v ¬∑ w = v‚ÇÅw‚ÇÅ + v‚ÇÇw‚ÇÇ + v‚ÇÉw‚ÇÉ = ||v|| ||w|| cos(Œ∏)

### üì¶ Real-World Example: Work in Physics
When you push a heavy box, the work done depends on both the force you apply AND how much that force aligns with the direction of movement. Push straight forward = maximum work. Push sideways = zero work (you're not helping the box move forward at all!).

**Key Properties**:
- If v ¬∑ w = 0, vectors are perpendicular (90¬∞ angle)
- If v ¬∑ w > 0, angle < 90¬∞ (pointing somewhat same direction)
- If v ¬∑ w < 0, angle > 90¬∞ (pointing somewhat opposite directions)

### üéØ Use Cases:
- **Machine Learning**: Measuring similarity between data points
- **Computer Graphics**: Lighting calculations (surface angle vs light direction)
- **Recommendation Systems**: Finding similar users/products
- **Physics**: Calculating work done by forces

```python
def angle_between(v, w):
    cos_angle = np.dot(v, w) / (np.linalg.norm(v) * np.linalg.norm(w))
    return np.arccos(np.clip(cos_angle, -1, 1))

# Physics example: pushing a box
force_applied = np.array([10, 3])      # Force: 10N forward, 3N upward
movement_direction = np.array([1, 0])   # Box moves purely forward

work_done = np.dot(force_applied, movement_direction)
angle = angle_between(force_applied, movement_direction)

print(f"Force applied: {force_applied} N")
print(f"Movement direction: {movement_direction}")
print(f"Work done (dot product): {work_done} (only forward component counts!)")
print(f"Angle between force and movement: {np.degrees(angle):.1f}¬∞")

# Similarity example (like recommendation systems)
# user1_preferences = np.array([5, 3, 1, 4])  # Ratings for movies A,B,C,D
# user2_preferences = np.array([4, 4, 2, 5])  # Another user's ratings
user1_preferences = np.array([1,1])  # Ratings for movies A,B,C,D
user2_preferences = np.array([-1,-1])  # Another user's ratings

similarity = np.dot(user1_preferences, user2_preferences)
print(f"\nUser 1 likes: {user1_preferences}")
print(f"User 2 likes: {user2_preferences}")
print(f"Similarity score: {similarity} (higher = more similar tastes)")
print(f"Angle between user1 and user2: {angle_between(user1_preferences, user2_preferences)}¬∞")
```

## 7. Cross Product (Vector Product)

**Formula**: v √ó w = (v‚ÇÇw‚ÇÉ - v‚ÇÉw‚ÇÇ, v‚ÇÉw‚ÇÅ - v‚ÇÅw‚ÇÉ, v‚ÇÅw‚ÇÇ - v‚ÇÇw‚ÇÅ)

### üîß Real-World Example: Torque (Twisting Force)
When you use a wrench to tighten a bolt, the twisting force (torque) depends on both the force you apply and the direction. The cross product gives you the axis of rotation and the strength of the twist. This is why longer wrenches make it easier to turn bolts!

**Properties**:
- Result is perpendicular to both input vectors
- Magnitude = ||v|| ||w|| sin(Œ∏)
- Right-hand rule determines direction

### üéØ Use Cases:
- **Engineering**: Calculating torque and rotational forces
- **Computer Graphics**: Finding surface normals for 3D lighting
- **Physics**: Angular momentum, magnetic fields
- **Robotics**: Determining rotation axes

```python
# Wrench/torque example
wrench_arm = np.array([0.3, 0, 0])    # 30cm wrench pointing right
applied_force = np.array([0, 20, 0])   # 20N force applied upward

torque = np.cross(wrench_arm, applied_force)
torque_magnitude = np.linalg.norm(torque)

print(f"Wrench arm: {wrench_arm} m")
print(f"Applied force: {applied_force} N")
print(f"Resulting torque: {torque} N‚ãÖm")
print(f"Torque magnitude: {torque_magnitude} N‚ãÖm")
print(f"Direction: {'clockwise' if torque[2] < 0 else 'counterclockwise'} (looking down z-axis)")

# Surface normal example (computer graphics)
# Three points defining a triangle surface
point1 = np.array([0, 0, 0])
point2 = np.array([1, 0, 0])
point3 = np.array([0, 1, 0])

# Two edges of the triangle
edge1 = point2 - point1
edge2 = point3 - point1

# Surface normal (perpendicular to the surface)
surface_normal = np.cross(edge1, edge2)
unit_normal = surface_normal / np.linalg.norm(surface_normal)

print(f"\nTriangle edges: {edge1}, {edge2}")
print(f"Surface normal: {surface_normal}")
print(f"Unit normal: {unit_normal}")
print("This normal vector is used for lighting calculations in 3D graphics!")
```

## 8. Vector Projection

**Scalar projection**: comp_w(v) = (v ¬∑ w) / ||w||

**Vector projection**: proj_w(v) = ((v ¬∑ w) / ||w||¬≤) √ó w

### üåû Real-World Example: Shadow Length
When sunlight hits a flagpole, the shadow on the ground is the projection of the pole onto the ground. The projection shows how much of the pole "points" in the ground's direction. At noon, the shadow is shortest because the pole is most perpendicular to the ground!

### üéØ Use Cases:
- **Computer Graphics**: Shadow rendering and lighting
- **Physics**: Component of force in a specific direction
- **Data Analysis**: Reducing dimensions while preserving important information
- **Engineering**: Finding components of forces along specific axes

```python
def vector_projection(v, w):
    return (np.dot(v, w) / np.dot(w, w)) * w

def scalar_projection(v, w):
    return np.dot(v, w) / np.linalg.norm(w)

# Shadow example
flagpole = np.array([0, 0, 10])      # 10m tall flagpole
ground_direction = np.array([1, 0, 0]) # Ground extends in x-direction
sun_angle = np.array([1, 0, 2])      # Sun at an angle

# Project flagpole onto ground to find shadow
shadow_vector = vector_projection(flagpole, ground_direction)
shadow_length = scalar_projection(flagpole, ground_direction)

print(f"Flagpole: {flagpole} m")
print(f"Shadow vector: {shadow_vector} m")
print(f"Shadow length: {shadow_length} m")

# Force component example
applied_force = np.array([50, 30])    # Force applied at an angle
desired_direction = np.array([1, 0])   # We only care about horizontal component

useful_force = vector_projection(applied_force, desired_direction)
useful_magnitude = scalar_projection(applied_force, desired_direction)

print(f"\nApplied force: {applied_force} N")
print(f"Useful horizontal force: {useful_force} N")
print(f"Useful magnitude: {useful_magnitude} N")
print(f"Wasted vertical component: {applied_force[1]} N")
```

## 9. Distance Between Points

**Formula**: d = ||P‚ÇÇ - P‚ÇÅ||

### üì± Real-World Example: GPS Distance
Your phone calculates the straight-line distance between two locations by treating them as points in space and finding the magnitude of the vector between them. This is the "as the crow flies" distance, different from driving distance!

### üéØ Use Cases:
- **Navigation**: "As the crow flies" distance calculations
- **Gaming**: Collision detection, AI pathfinding
- **Data Science**: Clustering similar data points
- **Logistics**: Optimizing delivery routes

```python
# GPS example (simplified 2D coordinates)
home = np.array([0, 0])        # Your home
work = np.array([3, 4])        # Your workplace
store = np.array([1, 7])       # Grocery store

home_to_work = np.linalg.norm(work - home)
home_to_store = np.linalg.norm(store - home)
work_to_store = np.linalg.norm(store - work)

print(f"Home location: {home}")
print(f"Work location: {work}")
print(f"Store location: {store}")
print(f"\nDistances:")
print(f"Home to work: {home_to_work:.1f} km")
print(f"Home to store: {home_to_store:.1f} km")
print(f"Work to store: {work_to_store:.1f} km")

# Gaming example - which enemy is closest?
player_pos = np.array([10, 15])
enemy1_pos = np.array([12, 18])
enemy2_pos = np.array([8, 12])
enemy3_pos = np.array([15, 20])

distances = {
    'Enemy 1': np.linalg.norm(enemy1_pos - player_pos),
    'Enemy 2': np.linalg.norm(enemy2_pos - player_pos),
    'Enemy 3': np.linalg.norm(enemy3_pos - player_pos)
}

closest_enemy = min(distances, key=distances.get)
print(f"\nPlayer at: {player_pos}")
print(f"Enemy distances: {distances}")
print(f"Closest enemy: {closest_enemy} at {distances[closest_enemy]:.1f} units away")
```

## 10. Linear Independence

Vectors are linearly independent if no vector can be written as a linear combination of others.

### üó∫Ô∏è Real-World Example: Unique Directions
Think of it like giving directions: "Go north, then east" uses two independent directions. But "Go north, then northeast" isn't independent because northeast is just a combination of north and east. You need truly different directions to describe any location!

### üéØ Use Cases:
- **Engineering**: Ensuring structural supports work in different directions
- **Data Science**: Finding truly different features in datasets
- **Computer Graphics**: Defining coordinate systems
- **Physics**: Identifying independent degrees of freedom

```python
def are_linearly_independent(vectors):
    matrix = np.column_stack(vectors)
    rank = np.linalg.matrix_rank(matrix)
    return rank == len(vectors)

# Independent directions (like compass directions)
north = np.array([0, 1, 0])
east = np.array([1, 0, 0])
up = np.array([0, 0, 1])

independent_check1 = are_linearly_independent([north, east, up])
print(f"North, East, Up vectors: {independent_check1}")
print("These are independent - you need all three to describe any 3D position!")

# Dependent directions (northeast is just north + east)
north = np.array([0, 1])
east = np.array([1, 0])
northeast = np.array([1, 1])  # This is just north + east!

independent_check2 = are_linearly_independent([north, east, northeast])
print(f"\nNorth, East, Northeast vectors: {independent_check2}")
print("These are dependent - northeast = north + east")

# Practical example: Can these forces balance each other?
force1 = np.array([1, 0])    # 1N east
force2 = np.array([0, 1])    # 1N north  
force3 = np.array([2, 0])    # 2N east (just 2x force1!)

independent_check3 = are_linearly_independent([force1, force2, force3])
print(f"\nForces analysis: {independent_check3}")
if not independent_check3:
    print("Force 3 is just 2x Force 1 - not truly independent!")
    print("You can't balance forces in all directions with these three.")
```

## 12. Practice Problems with Real-World Context

### Problem 1: Navigation System
A drone needs to fly from point A(2, 3) to point B(8, 7). Calculate:
- The displacement vector
- The distance it needs to travel
- The unit vector showing the direction

```python
# Drone navigation problem
point_A = np.array([2, 3])
point_B = np.array([8, 7])

displacement = point_B - point_A
distance = np.linalg.norm(displacement)
direction = unit_vector(displacement)

print(f"Starting point A: {point_A}")
print(f"Destination B: {point_B}")
print(f"Displacement vector: {displacement}")
print(f"Distance to travel: {distance:.2f} units")
print(f"Direction (unit vector): {direction}")
print(f"Direction in degrees: {np.degrees(np.arctan2(direction[1], direction[0])):.1f}¬∞ from east")
```

### Problem 2: Force Analysis
Two forces are acting on an object: F‚ÇÅ = (15, 20) N and F‚ÇÇ = (-10, 5) N.
Find the resultant force and determine if a third force F‚ÇÉ = (-5, -25) N would balance the system.

```python
# Force analysis problem
F1 = np.array([15, 20])   # Force 1
F2 = np.array([-10, 5])   # Force 2
F3 = np.array([-5, -25])  # Proposed balancing force

resultant_F1_F2 = F1 + F2
total_force = F1 + F2 + F3
resultant_magnitude = np.linalg.norm(resultant_F1_F2)

print(f"Force 1: {F1} N")
print(f"Force 2: {F2} N")
print(f"Resultant of F1 + F2: {resultant_F1_F2} N")
print(f"Magnitude of resultant: {resultant_magnitude:.2f} N")
print(f"\nProposed balancing force F3: {F3} N")
print(f"Total force (F1 + F2 + F3): {total_force} N")

if np.allclose(total_force, [0, 0], atol=1e-10):
    print("‚úÖ The system is balanced! F3 perfectly balances F1 and F2.")
else:
    print("‚ùå The system is not balanced.")
    needed_force = -(F1 + F2)
    print(f"Force needed for balance: {needed_force} N")
```

### Problem 3: Work and Energy
A person pushes a box with force F = (50, 30) N, and the box moves along displacement d = (4, 0) m.
Calculate the work done and explain why the vertical component doesn't contribute.

```python
# Work calculation problem
force = np.array([50, 30])      # Applied force in N
displacement = np.array([4, 0])  # Movement in m (purely horizontal)

work_done = np.dot(force, displacement)
force_magnitude = np.linalg.norm(force)
displacement_magnitude = np.linalg.norm(displacement)
angle = angle_between(force, displacement)

# Component analysis
horizontal_force = force[0]
vertical_force = force[1]

print(f"Applied force: {force} N")
print(f"  - Horizontal component: {horizontal_force} N")
print(f"  - Vertical component: {vertical_force} N")
print(f"Displacement: {displacement} m (purely horizontal)")
print(f"\nWork done: {work_done} J")
print(f"Angle between force and displacement: {np.degrees(angle):.1f}¬∞")
print(f"\nAlternative calculation: {force_magnitude:.1f} N √ó {displacement_magnitude} m √ó cos({np.degrees(angle):.1f}¬∞) = {force_magnitude * displacement_magnitude * np.cos(angle):.1f} J")
print(f"\nüí° Only the horizontal force component ({horizontal_force} N) does work.")
print(f"   The vertical component ({vertical_force} N) is perpendicular to motion, so it contributes 0 work.")
```

### Problem 4: 3D Torque Calculation
A wrench extends 0.25 m in the x-direction. A force of 40 N is applied in the y-direction.
Calculate the torque and determine the axis of rotation.

```python
# 3D Torque problem
wrench_vector = np.array([0.25, 0, 0])  # 25cm wrench in x-direction
applied_force = np.array([0, 40, 0])    # 40N force in y-direction

torque_vector = np.cross(wrench_vector, applied_force)
torque_magnitude = np.linalg.norm(torque_vector)

print(f"Wrench position vector: {wrench_vector} m")
print(f"Applied force: {applied_force} N")
print(f"Torque vector: {torque_vector} N‚ãÖm")
print(f"Torque magnitude: {torque_magnitude} N‚ãÖm")

# Determine rotation axis and direction
if torque_vector[2] > 0:
    rotation_direction = "counterclockwise (when viewed from positive z-axis)"
elif torque_vector[2] < 0:
    rotation_direction = "clockwise (when viewed from positive z-axis)"
else:
    rotation_direction = "no rotation about z-axis"

print(f"\nRotation axis: z-axis (pointing out of xy-plane)")
print(f"Rotation direction: {rotation_direction}")
print(f"\nüîß This is why longer wrenches make it easier to turn bolts!")
print(f"   Torque = distance √ó force √ó sin(angle)")
print(f"   Here: {torque_magnitude} = {wrench_vector[0]} √ó {applied_force[1]} √ó sin(90¬∞)")
```
